```python
from math import sqrt
import cmath
import sympy
import matplotlib.pyplot as plt
import numpy as np     
from sympy.solvers import solve
from sympy import Symbol
```

```python
a = -2
b = 4
c = 7
global a
global b
global c
```

## Определить корни

```python
x = Symbol('x')
result = solve(-2*x**2 + 4*x + 7, x)

print(f'Корни уравнения:\nx1 = {result[0]}\nx2 = {result[1]}')
```

## Найти интервалы, на которых функция возрастает

```python
x0 = -b/(a*2)

print(f'{x0} => +∞ ')

x = np.arange(-10,x0)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

plt.plot(x,-2*x**2 + 4*x + 7, 'r')
plt.grid()
plt.show()
```

## Найти интервалы, на которых функция убывает

```python
print(f'-∞ => {x0}')

x = np.arange(x0,10)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)


plt.plot(x,-2*x**2 + 4*x + 7, 'b')
plt.grid()
plt.show()
```

## Построить график

```python
minX = (x0-10)
maxX = x0 + 10

x = np.arange(minX,maxX)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

plt.plot(x,-2*x**2 + 4*x + 7, 'g')
plt.grid()
plt.show()
```

## Вычислить вершину

```python
x0 = -b/(a*2)
y0 = -2*x0**2 + 4*x0 + 7
print(f'Вершина функции: \n[{x0} ; {y0}]')
minX = x0-10
maxX = x0 + 10

x = np.arange(minX,maxX,1)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(1, 9,'ro')
plt.plot(x,-2*x**2 + 4*x + 7, 'g')

plt.grid()
plt.show()
```
